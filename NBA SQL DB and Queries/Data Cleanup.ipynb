```python
import pandas as pd

players_file = 'Players_Per100.csv'
teams_file = 'Teams_Stats_Per100.csv'

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 160)


players = pd.read_csv(players_file)
teams = pd.read_csv(teams_file)


print('players shape:', players.shape)
print('teams shape:', teams.shape)
print('')


print('PLAYERS - .info()')
players.info()

print('PLAYERS - .dtypes')
print(players.dtypes)

print('PLAYERS - .describe() numeric')
print(players.describe())

print('PLAYERS - .describe() object/strings')
print(players.describe(include='object'))

print('PLAYERS - .isnull().sum()')
print(players.isnull().sum())

print('PLAYERS - .value_counts() on a few columns')

print('season')
print(players['season'].value_counts())

print('league')
print(players['league'].value_counts())

print('position')
print(players['position'].value_counts(dropna=False))

print('team (top 15)')
print(players['team'].value_counts().head(15))


print('TEAMS - .info()')
teams.info()

print('TEAMS - .dtypes')
print(teams.dtypes)

print('TEAMS - .describe() numeric')
print(teams.describe())

print('TEAMS - .describe() object/strings')
print(teams.describe(include='object'))

print('TEAMS - .isnull().sum()')
print(teams.isnull().sum())

print('TEAMS - .value_counts() on a few columns')


print('season')
print(teams['season'].value_counts())

print('league')
print(teams['league'].value_counts())

print('playoffs')
print(teams['playoffs'].value_counts())

print('abbreviation (top 15)')
print(teams['abbreviation'].value_counts().head(15))
```
